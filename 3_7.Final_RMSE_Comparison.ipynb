```python
import matplotlib.pyplot as plt

# Sample dictionary with model names keys and RMSE values
data = {
        'KNN': 1.29, 
        'NMF': 1.29, 
        'ANN': 0.11, 
        'Emb_Reg.': 0.82, 
        'Emb_Clas.': 1.29
    }

# Extract keys and values from the dictionary
categories = list(data.keys())
values = list(data.values())

# Create the bar plot
plt.bar(categories, values, color=['steelblue', 'teal', 'mediumaquamarine', 'lightseagreen', 'skyblue'] ) 

# Add labels and title for clarity
plt.ylabel('RMSE')
plt.title('Model RMSE comparison')

# Display the plot
plt.show()
```
